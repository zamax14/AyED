### a) ¿Quién es la población y la muestra?

### d) ¿Hay evidencia para afirmar que la Planta A tiene mejor clima laboral? Para contestar esta pregunta haga un análisis descriptivo comparativo de los datos (medidas de tendencia central, de variabilidad y obtenga el histograma).

**Respuesta:**


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
planta_a = np.array([86, 85, 100, 88, 94, 93, 85, 74, 95, 90, 
                     86, 93, 86, 93, 75, 72, 85, 85, 78, 75, 
                     99, 96, 84, 82, 79, 84, 79, 89, 84, 75, 
                     75, 83, 75, 77, 89, 84, 76, 75, 91, 84, 
                     91, 85, 96, 55, 83, 97, 64, 93, 82, 92])

planta_b = np.array([83, 73, 83, 87, 78, 84, 83, 91, 78, 79, 
                     81, 88, 83, 80, 80, 80, 80, 82, 81, 82, 
                     82, 85, 77, 84, 85, 81, 75, 77, 84, 76, 
                     78, 87, 79, 73, 88, 87, 81, 85, 80, 80, 
                     86, 75, 78, 79, 75, 79, 83, 82, 75, 74])
```

```python
print("Planta A - Media:", np.mean(planta_a))
print("Planta A - Mediana:", np.median(planta_a))
print("Planta A - Moda:", stats.mode(planta_a).mode)
print("Planta A - Rango:", np.max(planta_a) - np.min(planta_a))
print("Planta A - Varianza:", np.var(planta_a))
print("Planta A - Desviacion Estandar:", np.std(planta_a))
print("Planta A - Coeficiente de Variacion:", np.std(planta_a)/np.mean(planta_a))

n_bins = int(np.ceil(1+np.log2(len(planta_a))))

plt.hist(planta_a, bins=n_bins)
plt.title("Histograma Planta A")
plt.show()
```

```python
print("Planta B - Media:", np.mean(planta_b))
print("Planta B - Mediana:", np.median(planta_b))
print("Planta B - Moda:", stats.mode(planta_b).mode)
print("Planta B - Rango:", np.max(planta_b) - np.min(planta_b))
print("Planta B - Varianza:", np.var(planta_b))
print("Planta B - Desviacion Estandar:", np.std(planta_b))
print("Planta B - Coeficiente de Variacion:", np.std(planta_b)/np.mean(planta_b))

n_bins = int(np.ceil(1+np.log2(len(planta_b))))
    
plt.hist(planta_b, bins=6)
plt.title("Histograma Planta B")
plt.show()
```

Según los datos vistos, podemos ver que la media y mediana de 'A' son mayores a las de 'B', lo cual nos haría pensar que el clima laboral es mejor en 'A', pero según nuestras medidas de dispersión, podemos ver que los valores de 'B' tienen una variación menor que la de 'A', esto se puede apreciar de mejor manera en los histogramas al ver que la planta 'B' tiene una distribución normal más simétrica que en el caso de 'A'.
Con esto podemos concluir que pese a que la planta 'A' tiene valores muy altos de satisfacción, también tiene un mayor número de personas insatisfechas a comparacion de 'B', mientras que en 'B' la satisfacción de los empleados está más balanceada y se puede llegar a considerar un mejor ambiente laboral.
