### Problema 1
**Un inspector de la Procuraduría Federal del Consumidor acude a una planta que elabora alimentos para verificar el cumplimiento de lo estipulado en los envases de los productos en cuanto a peso y volumen. Uno de los productos que decide analizar es el peso de las cajas de cereal, en las cuales, para una de sus presentaciones, se establece que el contenido neto es de 300 gramos, y sospecha que están entregando menos. El inspector toma una muestra de 25 cajas y pesa su contenido. La media y desviación estándar de la muestra son = 298.3 y S = 4.5.**

- a) Formula la hipótesis nula y alternativa apropiada al problema.
- b) Pruebe la hipótesis que formuló con un nivel de significancia del 5%. Interprete.
- c) Estime mediante un intervalo al 95% confianza el verdadero valor de $\mu$.
- d) ¿Cuál es el error de estimación en el inciso anterior? Explique.

#### Inciso A

**Hipótesis:**
- **H₀: μ = 300** (El peso promedio es 300g)
- **H₁: μ < 300** (El peso promedio es menor a 300g)

Es una prueba de cola izquierda porque se sospecha que pesan menos.

```python
import math

x_bar = 298.3
s = 4.5
n = 25
mu_0 = 300
alpha = 0.05
```

#### Inciso B

**Prueba t (σ desconocida, n pequeño)**

Estadístico: $t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$

```python
t_calculado = (x_bar - mu_0) / (s / math.sqrt(n))
t_critico = -1.711

print(f"t calculado = {t_calculado:.4f}")
print(f"t crítico = {t_critico}")
```

#### Inciso C

Intervalo de confianza 95%

Fórmula: $\bar{x} \pm t_{\alpha/2} \cdot \frac{s}{\sqrt{n}}$

```python
t_alpha = 2.064
error_estandar = s / math.sqrt(n)
margen_error = t_alpha * error_estandar

limite_inferior = x_bar - margen_error
limite_superior = x_bar + margen_error

print(f"Error estandar = {error_estandar:.4f}")
print(f"Margen de error = {margen_error:.4f}")
print(f"Intervalo de confianza: [{limite_inferior:.2f}, {limite_superior:.2f}]")
```

#### Inciso D

```python
print(f"Error de estimación = {margen_error:.4f} gramos")
```
